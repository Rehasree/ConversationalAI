```python
!pip install openai python-dotenv pygame

import openai
import os
from google.colab import files
from openai import OpenAI
import io
from IPython.display import Audio, display,HTML
import base64

# Set up OpenAI API key
openai.api_key = "sk-proj-ssPAAPHJ79tSIGWaPUlMT3BlbkFJODIJnhTQzk3MTlab5utY"  # Replace with your actual API key

# Initialize the OpenAI client
client = OpenAI(api_key="sk-proj-ssPAAPHJ79tSIGWaPUlMT3BlbkFJODIJnhTQzk3MTlab5utY")


def upload_audio():
    uploaded = files.upload()
    if uploaded:
        filename = next(iter(uploaded))
        return filename
    return None

def speech_to_text(filename):
    with open(filename, "rb") as audio_file:
        transcript = client.audio.transcriptions.create(
            model="whisper-1",
            file=audio_file
        )
    return transcript.text

def get_gpt_response(prompt, conversation_history):
    system_message = """
    You are a helpful AI assistant for a hotel booking system. Respond to user queries about booking rooms and hotel amenities.
    Available room types and prices are:
    1. Single Room (USD 100 per day)
    2. Double Room (USD 150 per day)
    3. Deluxe Room (USD 200 per day)
    4. Suite (USD 175 per day)
    5. Family Room (USD 275 per day)

    When a user selects a room, calculate the total price for their stay and provide a brief description of the room.
    If asked about dining options, mention that the hotel has breakfast, lunch, dinner, and in-room dining options.
    """

    messages = [
        {"role": "system", "content": system_message},
        *conversation_history,
        {"role": "user", "content": prompt}
    ]

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages
    )
    return response.choices[0].message.content

def text_to_speech(text):
    response = client.audio.speech.create(
        model="tts-1",
        voice="nova",
        input=text
    )

    audio_base64 = base64.b64encode(response.content).decode('utf-8')

    audio_html = f'''
    <audio controls autoplay>
        <source src="data:audio/mp3;base64,{audio_base64}" type="audio/mp3">
        Your browser does not support the audio element.
    </audio>
    '''

    display(HTML(audio_html))

def hotel_booking_bot():
    print("Welcome to the Hotel Booking AI Assistant!")
    conversation_history = []

    while True:
        print("\nPlease upload an audio file with your booking request or question.")
        audio_file = upload_audio()

        if audio_file:
            user_input = speech_to_text(audio_file)
            print(f"You said: {user_input}")

            if user_input.lower() in ["exit", "quit", "bye", "stop"]:
                print("Thank you for using our booking service. Goodbye!")
                break

            conversation_history.append({"role": "user", "content": user_input})
            bot_response = get_gpt_response(user_input, conversation_history)
            conversation_history.append({"role": "assistant", "content": bot_response})

            print(f"AI: {bot_response}")

            print("AI response as audio:")
            text_to_speech(bot_response)
        else:
            print("No file uploaded. Please try again.")

if __name__ == "__main__":
    hotel_booking_bot()


```
